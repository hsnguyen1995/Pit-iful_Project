```python
# Import dependencies
import pandas as pd

# Read and load csv
avocado_path = "../Data/avocado.csv"
avo_csv = pd.read_csv(avocado_path)

avo_csv
```

```python
# check data types
avo_csv.dtypes
```

```python
# Check for NaNs
avo_csv.count()
```

```python
# Extract month from date column
monthList = []
monthExtract = avo_csv["Date"].str.split("/").str[1]

# Remove leading 0's in month values because of inconsistent month format
for i in monthExtract:
    while i[0] == "0":
        i = i[1:]
    monthList.append(i)
```

```python
# Load months column into existing DF
avo_csv["Month"] = monthList

avo_csv
```

```python
# Convert Year column to string and concatenate with Month column to get Month/Year column
avo_csv["Year"] = avo_csv["Year"].astype(str)
avo_csv["Month/Year"] = avo_csv["Month"] + "/" + avo_csv["Year"]

avo_csv.columns
```

```python
# Rearrange columns, drop unnecessary columns
final_avo_df = avo_csv[['Month/Year', 'AveragePrice', 'Total Volume', 'Small Hass', 'Large Hass', 'XL Hass', 'Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags', 'Type', 'Region']]
final_avo_df
```

```python
# Export DF as csv
final_avo_df.to_csv("../Data/avocado_clean.csv", index=False, header=True)
```
