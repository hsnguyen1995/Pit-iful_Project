```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_avocado_diseases'
```

```python
tables = pd.read_html(url)
tables
```

```python
type(tables)
```

```python
bacterialdf = tables[0]
bacterialdf.columns=['Common Name', 'Scientific Name']
bacterialdf['Common Name']=bacterialdf['Common Name'].astype(str) + ' (Bacteria)'
bacterialdf
```

```python
fungaldf = tables[1]
fungaldf.columns=['Common Name', 'Scientific Name']
fungaldf['Common Name']=fungaldf['Common Name'].astype(str) + ' (Fungus)'
fungaldf
```

```python
virusdf = tables[2]
virusdf.columns=['Common Name', 'Cause']
virusdf['Common Name']=virusdf['Common Name'].astype(str) + ' (Virus)'
virusdf
```

```python
miscdf = tables [3]
miscdf.columns=['Common Name', 'Cause']
miscdf['Common Name']=miscdf['Common Name'].astype(str) + ' (Misc.)'
miscdf
```

```python
frames = [bacterialdf, virusdf, fungaldf, miscdf]
combined = pd.concat(frames, sort=False)
combined
```

```python
combined.to_html('avocado_diseases.html')
```

```python
combined.to_csv('avocado_diseases.csv')
```
