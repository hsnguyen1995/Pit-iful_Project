```python
import pandas as pd
```

```python
url = 'https://www.californiaavocadogrowers.com/industry/industry-statistical-data'
```

```python
tables=pd.read_html(url)
tables
```

```python
type(tables)
```

```python
annualsince71df = tables[0]
annualsince71df.columns=['Year', 'Bearing Acres', 'Volume (millions of pounds)', 
                         'Crop Value ($)', 'Price Per LB',  
                         'Average $ per Acre', 'Average Lbs Per Acre']
annualsince71df
```

```python
reduced71df=annualsince71df.drop(['Volume (millions of pounds)', 
                         'Crop Value ($)', 'Price Per LB'], axis=1)
reduced71df
```

```python
reduced71df.to_csv('annualsince71.csv')
```

```python
reduced71df.to_html('annualsince71.html')
```
